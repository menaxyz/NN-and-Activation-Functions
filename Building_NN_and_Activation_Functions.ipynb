1. Build a NN with structure as described below:
<br>Input layer: size 1
<br>Hidden layer 1: 8 neurons, dropout 0.5, densely connected
<br>Hidden layer 2: 3 neuron, densely connected
<br>Output layer: size 1

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential([
    Dense(8, input_dim=1),
    Dropout(0.5),
    Dense(3),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()
```

Activation Functions
The model you build above does not have any activation function.
Now, try to add the same data that is generated in file Introduction to Artificial Neural Network to this file.
Add activation functions to you model, and train your model using different activation functions/combination of activation functions.

You need to try at least 5 different cases.

```python
# case1 sigmiod
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
import numpy as np
import matplotlib.pyplot as plt

model = Sequential([
    Dense(8, input_dim=1, activation='sigmoid'),
    Dropout(0.5),
    Dense(3, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])


model.compile(optimizer='adam', loss='mean_squared_error')


x = np.linspace(-5, 5, 100)
y = model.predict(x)


plt.plot(x, y)
plt.title("sigmoid Activation")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()


model.summary()
```

```python
# case2 tanh
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
import numpy as np
import matplotlib.pyplot as plt

model = Sequential([
    Dense(8, input_dim=1, activation='tanh'),
    Dropout(0.5),
    Dense(3, activation='tanh'),
    Dense(1, activation='tanh')
])

model.compile(optimizer='adam', loss='mean_squared_error')

x = np.linspace(-5, 5, 100)
y = model.predict(x)


plt.plot(x, y)
plt.title("tanh Activation")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()



model.summary()
```

```python
# case5 LeakyRelu
import tensorflow as tf
from tensorflow.keras import Sequential
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense, Dropout
import numpy as np


leaky_relu = tf.keras.layers.LeakyReLU(alpha=0.01)
model = Sequential([
    Dense(8, input_dim=1, activation='leaky_relu'),
    Dropout(0.5),
    Dense(3, activation='leaky_relu'),
    Dense(1, activation='leaky_relu')
])



model.compile(optimizer='adam', loss='mean_squared_error')

x = np.linspace(-5, 5, 100)
y = model.predict(x)


plt.plot(x, y)
plt.title("Leaky ReLU Activation")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()

model.summary()
```
